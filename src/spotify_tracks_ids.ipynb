```python
import pandas as pd

df_train = pd.read_csv("../data/music-genre-classification/train.csv")
df_train
```

```python
from pathlib import Path


def save_df_as_csv(df: pd.DataFrame, path: str):
    filepath = Path(path)
    filepath.parent.mkdir(parents=True, exist_ok=True)
    df.to_csv(filepath, index=False)


spotify_id_key = "spotify_id"
artist_name_key = "artist_name"
track_name_key = "track_name"
class_key = "class"

dic_spotify_ids_tracks = {
    spotify_id_key: [],
    artist_name_key: [],
    track_name_key: [],
    class_key: [],
}

df_spotify_ids = pd.DataFrame(dic_spotify_ids_tracks)
file_path = "../data/spotify-tracks-ids.csv"

try:
    df_spotify_ids = pd.read_csv(file_path)
except:
    save_df_as_csv(df_spotify_ids, file_path)

df_spotify_ids
```

```python
artist_name_df_key = "Artist Name"
track_name_df_key = "Track Name"
class_df_key = "Class"

df_train_batch = df_train[[artist_name_df_key, track_name_df_key, class_df_key]]
df_train_batch.rename(
    columns={
        artist_name_df_key: artist_name_key,
        track_name_df_key: track_name_key,
        class_df_key: class_key,
    },
    inplace=True,
)
df_spotify_ids_batch = df_spotify_ids[[artist_name_key, track_name_key, class_key]]
df_batched = pd.concat([df_train_batch, df_spotify_ids_batch]).drop_duplicates(
    keep=False
)
df_batched
```

```python
from service.spotify_service import SpotifyService

spotify_service = SpotifyService()

dic_spotify_ids_tracks = {
    spotify_id_key: [],
    artist_name_key: [],
    track_name_key: [],
    class_key: [],
}

rows, columns = df_batched.shape

for i in range(rows):
    row = df_batched.iloc[i]
    a_name = row[0]
    t_name = row[1]
    t_class = row[2]
    spotify_id = pd.NA
    try:
        spotify_id = spotify_service.get_spotify_track_id(t_name, a_name)
    except Exception as ex:
        print(ex)
        break
    dic_spotify_ids_tracks[spotify_id_key].append(spotify_id)
    dic_spotify_ids_tracks[artist_name_key].append(a_name)
    dic_spotify_ids_tracks[track_name_key].append(t_name)
    dic_spotify_ids_tracks[class_key].append(t_class)
```

```python
df_spotify_tracks_ids = pd.DataFrame(dic_spotify_ids_tracks)
df_spotify_tracks_ids
```

```python
df_tracks_appended = df_spotify_tracks_ids.append(df_spotify_ids)
save_df_as_csv(df_tracks_appended, "../data/spotify-tracks-ids.csv")
```
